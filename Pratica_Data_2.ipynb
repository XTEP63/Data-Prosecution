```python
import pandas as pd 
import os 
```

```python
Datas = os.listdir("./Data/")
Datas
```

```python
def archive_type(path):
    
    dotIndex = path.rfind(".")
    extencion = path[dotIndex + 1 :]
    return extencion
```

```python
for i in range(len(Datas)):
    extenxion = archive_type(Datas[i])
    print(extenxion)
    if extenxion == "cvs":
        df = pd.read_csv("./Data/"  + Datas[i])
    elif extenxion == "json":
        df = pd.read_json("./Data/"  + Datas[i])
    
    print(df.head(5))
    print(df.tail(5))
    print(df.columns)
    print(df.iloc[:, 1])
    print(df.info())
    print(df.describe())
    print(df.shape) 
    print(df.max())
    print(df.min())
    print(df.cumsum())
```
